```python
def extract_data_by_1000_for_site(data_folder, cell, dataset):

    file_paths = glob.glob(os.path.join(data_folder, "*.hdf5"))
    paths = []
    
    for file_path in file_paths:
        with h5py.File(file_path, "r") as file:
            outputs = file[dataset + "_grids"]
            if np.count_nonzero(outputs) == 0:
                print("all zeros")
                continue
            
            for i in range(outputs.shape[0]):
                path = outputs[i, :, 105-cell[0], cell[1], :, :]
                paths.append(path)
    return paths
```
